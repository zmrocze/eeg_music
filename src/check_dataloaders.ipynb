```python
from data import copy_from_dataloader_into_dir


def create_combined_training_calibration_dataset(
  bcmi_root_path: str, new_dataset_save_dir: Path
) -> EEGMusicDataset:
  """Create an EEGMusicDataset that combines BCMI training and calibration data."""

  training_loader = BCMITrainingLoader(f"{bcmi_root_path}/bcmi-training")
  training_loader.load_all_subjects()
  copy_from_dataloader_into_dir("training", training_loader, new_dataset_save_dir)
  del training_loader
  calibration_loader = BCMICalibrationLoader(f"{bcmi_root_path}/bcmi-calibration")
  calibration_loader.load_all_subjects()
  copy_from_dataloader_into_dir("calibration", calibration_loader, new_dataset_save_dir)
  del calibration_loader

  combined_dataset = EEGMusicDataset.load_ondisk(new_dataset_save_dir)

  return combined_dataset
```

```python
# Create the combined training and calibration dataset
from pathlib import Path

temp_save_dir = Path("/tmp/bcmi_combined_dataset")

combined_dataset = create_combined_training_calibration_dataset(
  "/home/zmrocze/studia/uwr/magisterka/datasets/bcmi", temp_save_dir
)

print(f"Combined dataset: {len(combined_dataset.df)} trials")
print(f"Datasets: {combined_dataset.df['dataset'].value_counts().to_dict()}")
combined_dataset.df.head()
```

```python
# Access trial data from the new structure
first_trial = combined_dataset.df.iloc[0]
print(f"Trial dataset: {first_trial['dataset']}")
print(f"Subject: {first_trial['subject']}")
print(f"Music reference: {first_trial['music_ref'].filename}")

# Get EEG data
eeg_data = first_trial["eeg_data"].get_eeg()
print(f"EEG type: {type(eeg_data)}")
print(f"EEG duration: {eeg_data.raw_eeg.times[-1]:.2f}s")

# Get music data from music collection
music_ref = None
for ref in combined_dataset.music_collection.keys():
  if ref.filename.filename == first_trial["music_ref"].filename:
    music_ref = ref
    break

if music_ref:
  music_data = combined_dataset.music_collection[music_ref].get_music()
  print(f"Music type: {type(music_data)}")
  print(f"Music duration: {len(music_data.raw_data) / music_data.sample_rate:.2f}s")
else:
  print("Music reference not found in collection")
```

```python
# Complete end-to-end test: reload the dataset from disk
reloaded_dataset = EEGMusicDataset.load_ondisk(temp_save_dir)

print(f"âœ… Success! Reloaded dataset from disk")
print(f"   - {len(reloaded_dataset.df)} trials")
print(f"   - {len(reloaded_dataset.music_collection)} music files")
print(f"   - Datasets: {reloaded_dataset.df['dataset'].value_counts().to_dict()}")

# Test accessing a trial
test_trial = reloaded_dataset.df.iloc[0]
test_eeg = test_trial["eeg_data"].get_eeg()
print(
  f"   - EEG data accessible: {type(test_eeg).__name__}, {test_eeg.raw_eeg.n_times} samples"
)

print("\nðŸŽ‰ All generic typing and dataset workflow tests passed!")
```

```python
# Verify remove_short_trials method
min_len = 10.0
before = len(reloaded_dataset)
filtered = reloaded_dataset.remove_short_trials(min_len)
after = len(filtered)
print(f"Filtered short trials: {before} -> {after} (threshold={min_len}s)")

# Sanity check: all remaining trials satisfy the constraint
ok = True
violations = 0
for i in range(len(filtered)):
  t = filtered[i]
  raw = t.eeg_data.get_eeg().raw_eeg
  sf = float(raw.info["sfreq"]) if "sfreq" in raw.info else raw.info["sfreq"]
  eeg_sec = raw.n_times / sf
  mus_sec = t.music_data.get_music().length_seconds()
  if eeg_sec < min_len or mus_sec < min_len:
    ok = False
    violations += 1
print("All constraints satisfied" if ok else f"Found {violations} violations")
```
